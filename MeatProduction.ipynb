```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv("/Users/kurt/Desktop/meat-production-tonnes.csv") 


# rename columns 
df1 = df.rename(columns = {'Livestock Primary - Meat, Total - 1765 - Production - 5510 - tonnes':'Meat_Production(tonnes)'})

df2 = df1[(df1['Year']>=2000)]
data= df2[(df2['Entity']=='Asia') |
                    (df2['Entity']=='Europe') |
                    (df2['Entity']=='Central America') | 
                    (df2['Entity']=='South America') | 
                    (df2['Entity']=='Northern America')|
          (df2['Entity']=='Oceania')]


data2 = data[['Entity','Year','Meat_Production(tonnes)']]

fig, ax = plt.subplots(figsize=(15,7))
# use unstack()
data2.groupby(['Year','Entity']).sum()['Meat_Production(tonnes)'].unstack().plot(ax=ax)



```

```python
data2.groupby(['Entity'])['Meat_Production(tonnes)'].sum().plot()
```

```python
data2.groupby(['Entity'])['Meat_Production(tonnes)'].sum().plot(kind='bar')
```

```python
data3 = data2[(data2['Year']>=2010)]
data3.groupby(['Entity', 'Year'])['Meat_Production(tonnes)'].sum().unstack().plot()

```

```python

fig, ax = plt.subplots()
data2.groupby('Entity').plot(x='Year', y='Meat_Production(tonnes)', ax=ax, legend=False)
```

```python
group = data2.groupby(["Entity","Year"]).sum()

total_price = group["Meat_Production(tonnes)"].groupby(level=0, group_keys=False)

total_price.nlargest(5).plot(kind="bar")
```

```python
data2.plot(x="Year", y="Meat_Production(tonnes)", kind="scatter")
```
